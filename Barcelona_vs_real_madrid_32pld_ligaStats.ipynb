# Basic Statistics of Real Madrid & Barcelona over 32 La Liga Matchdays With Python

```python
# import librarys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import seaborn as sns
```


## Data Source

- Real Madrid data : https://www.laliga.com/en-ES/clubs/real-madrid/stats

- FC Barcelona : https://www.laliga.com/en-ES/clubs/fc-barcelona/stats

```python
# load and display real madrid data
data_real = pd.read_csv("real_madrid_data.csv")
data_real
```

```python
# load and display fc barcelona data
data_barca = pd.read_csv("barcelona_data.csv")
data_barca
```

## General and Descriptive Statistics

```python
# Total Squad Size

real_players = data_real["NAME"].nunique()
barca_players = data_barca["NAME"].nunique()

squad_of_two = pd.DataFrame({
    'All players': [barca_players, real_players],
    'Team': ['Barcelona', 'Real Madrid']
    
})
squad_of_two
```

```python
# Position Distribution for Real madrid

dis_pos_real = data_real["POSITION"].value_counts().reset_index()
dis_pos_real.columns = ['POSITION', 'COUNT']
dis_pos_real
```

```python
# Position Distribution for Barcelona

dis_pos_barca = data_barca["POSITION"].value_counts().reset_index()
dis_pos_barca.columns = ['POSITION', 'COUNT']
dis_pos_barca
```

```python
# Team aggregate statistics of two clubs

real_global_stats = data_real[['GOALS','PEN', 'YC', 'RC']].sum()
real_global_stats_df = real_global_stats.reset_index()
real_global_stats_df.columns = ['Stat', 'Total']
real_global_stats_df
```

```python
barca_global_stats = data_barca[['GOALS','PEN', 'YC', 'RC']].sum()
barca_global_stats_df = barca_global_stats.reset_index()
barca_global_stats_df.columns = ['Stat', 'Total']
barca_global_stats_df
```

```python
# Average Statistics per Player in real madrid

real_mean = data_real[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].mean()
real_median = data_real[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].median()
real_min = data_real[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].min()
real_max = data_real[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].max()

real_mean_df = real_mean.reset_index()
real_mean_df.columns = ['Stat', 'Total']
real_mean_df
```

```python
# Average Statistics per Player in barcelona

barca_mean = data_barca[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].mean()
barca_median = data_barca[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].median()
barca_min = data_barca[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].min()
barca_max = data_barca[['MIN', 'PLD', 'GOALS', 'YC', 'RC']].max()

barca_mean_df = barca_mean.reset_index()
barca_mean_df.columns = ['Stat', 'Total']
barca_mean_df
```

### Offensive Performance Analysis

```python
barca_scorer = data_barca[['POSITION','NAME','PLD', 'GOALS']].sort_values(by='GOALS', ascending=False)
```

```python
# Top 5 scorers
barca_scorer.head()
```

```python
real_scorer = data_real[['POSITION','NAME','PLD', 'GOALS']].sort_values(by='GOALS', ascending=False)
```

```python
# Top 5 scorers
real_scorer.head()
```

```python
# Scorer top for two clubs

real = real_scorer.loc[real_scorer['POSITION'] == 'Forward'].copy()
real['TEAM'] = 'Real Madrid'
barca = barca_scorer.loc[barca_scorer['POSITION'] == 'Forward'].copy()
barca['TEAM'] = 'Barcelona'

two_scorers = pd.concat([real, barca])
two_scorers = two_scorers.sort_values(by='GOALS', ascending = False).reset_index(drop=True)
two_scorers.head(5)
```

```python
# Best forward of season
barca_forward = barca['GOALS'].sum()
real_forward = real['GOALS'].sum()

best_forrward = pd.DataFrame({
    'SUM OF GOALS': [barca_forward, real_forward],
    'TEAM': ['Barcelona', 'Real Madrid']
})
best_forrward
```

```python
# Best Goalkeeper
real_keeper = data_real.loc[(data_real['POSITION'] == 'Goalkeeper') & (data_real['PLD'] >= 3)].copy()
real_keeper['TEAM'] = 'Real Madrid'
barca_keeper = data_barca.loc[(data_barca['POSITION'] == 'Goalkeeper')  & (data_barca['PLD'] >= 3)].copy()
barca_keeper['TEAM'] = 'Barcelona'

two_goalkeeper = pd.concat([real_keeper, barca_keeper]).copy()
two_goalkeeper['GC/PLD'] =  two_goalkeeper['GC'] / two_goalkeeper['PLD']
two_goalkeeper = two_goalkeeper[['POSITION','NAME','GC','PLD', 'TEAM', 'GC/PLD']].sort_values(by='GC/PLD', ascending = True).reset_index(drop=True)
two_goalkeeper
```

```python
best_barca_scorer = data_barca.loc[data_barca['GOALS'].idxmax(), ['POSITION', 'NAME', 'GOALS', 'PEN']]
best_barca_scorer['NO PEN'] = best_barca_scorer['GOALS'] - best_barca_scorer['PEN']
best_real_scorer = data_real.loc[data_real['GOALS'].idxmax(), ['POSITION', 'NAME', 'GOALS', 'PEN']]
best_real_scorer['NO PEN'] = best_real_scorer['GOALS'] - best_real_scorer['PEN']
```

```python
# best socerers of the two teams 

best_socerer_real_vs_barca = pd.DataFrame({
    'NAME': [best_barca_scorer['NAME'], best_real_scorer['NAME']],
    'NO PEN': [best_barca_scorer['NO PEN'], best_real_scorer['NO PEN']],
    'PEN': [best_barca_scorer['PEN'], best_real_scorer['PEN']],
    'TEAM': ['Barcelona', 'Real Madrid']
}).set_index('NAME')

best_socerer_real_vs_barca
```

```python
# Bars chart
best_socerer_real_vs_barca.plot(kind="bar", stacked=True, color=['powderblue', 'red'])
plt.title('Ratio of penalty and no penaty goals')
plt.ylabel('Goals')
plt.show()
```

```python
# Real Scoring Efficiency per 90 Minutes

data_real.loc[data_real['POSITION'] == 'Forward', 'EFF_PER_90'] = data_real['GOALS'] / (data_real['MIN']) / 90
efficient_real_scorer = data_real[['POSITION', 'NAME','GOALS', 'EFF_PER_90']].sort_values(by='EFF_PER_90', ascending=False)
efficient_real_scorer.head(1)
```

```python
# Real Scoring Efficiency per Match
data_real.loc[data_real['POSITION'] == 'Forward', 'EFF_PER_MATCH'] = data_real['GOALS'] / data_real['PLD']
efficient_real_macth = data_real[['POSITION', 'NAME','GOALS', 'EFF_PER_MATCH']].sort_values(by='EFF_PER_MATCH', ascending=False)
efficient_real_macth.head(1)
```

```python
# Real Scoring Efficiency per 90 Minutes

data_barca.loc[data_barca['POSITION'] == 'Forward', 'EFF_PER_90'] = data_barca['GOALS'] / (data_barca['MIN']) / 90
efficient_barca_scorer = data_barca[['POSITION', 'NAME','GOALS', 'EFF_PER_90']].sort_values(by='EFF_PER_90', ascending=False)
efficient_barca_scorer.head(1)
```

```python
# Barca Scoring Efficiency per Match
data_barca.loc[data_barca['POSITION'] == 'Forward', 'EFF_PER_MATCH'] = data_barca['GOALS'] / data_barca['PLD']
efficient_barca_macth = data_barca[['POSITION', 'NAME','GOALS', 'EFF_PER_MATCH']].sort_values(by='EFF_PER_MATCH', ascending=False)
efficient_barca_macth.head(1)
```

```python
# Real Average Goals per Position
real_avg_pos = data_real.groupby('POSITION')['GOALS'].mean()
real_avg_pos_df = real_avg_pos.reset_index()
real_avg_pos_df.columns = ['POSITION', 'AVG GOALS/POSITION']
real_avg_pos_df.sort_values(by='AVG GOALS/POSITION', ascending=False)
```

```python
# Barca Average Goals per Position
barca_avg_pos = data_barca.groupby('POSITION')['GOALS'].mean()
barca_avg_pos_df = barca_avg_pos.reset_index()
barca_avg_pos_df.columns = ['POSITION', 'AVG GOALS/POSITION']
barca_avg_pos_df.sort_values(by='AVG GOALS/POSITION', ascending=False)
```
